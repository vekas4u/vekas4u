```python
x=data.drop(columns=['quality'])
y=data['quality']
```

```python
# Lets see how features are related to class
plt.figure(figsize=(15,30),facecolor='white')
plotnumber = 1

for column in x:
    if plotnumber<=12:
        ax = plt.subplot(6,2,plotnumber)
        sns.stripplot(y,x[column])
    plotnumber+=1
plt.tight_layout()
```

```python
#Great Lets proceed by checking multicollinearly in the dependent variables.Before that,we should 
# scale our data,Lets use the standard scalar for that.
```

```python
scalar = StandardScaler()
x_scaled = scalar.fit_transform(x)
```

```python
vif = pd.DataFrame()
vif['vif']=[variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vif["Features"] = x.columns

# Lets check the values
vif
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state = 355)
```

```python
log_reg = LogisticRegression()

log_reg.fit(x_train,y_train)
```

```python
y_pred = log_reg.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
# confusion matrix
conf_mat = confusion_matrix(y_test,y_pred)
conf_mat
```
