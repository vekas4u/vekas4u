```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lets see how data is distributed in every coloumn
plt.figure(figsize=(20,25),facecolor='white')
plotnumber=1

for column in data:
    if plotnumber<=12:   # as there are 12 columns in the data
        ax = plt.subplot(4,3,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
        
    plotnumber+=1
plt.show()
```

```python
x=data.drop(columns=['quality'])
y=data['quality']
```

```python
# Lets see how features are related to class
plt.figure(figsize=(15,30),facecolor='white')
plotnumber = 1

for column in x:
    if plotnumber<=12:
        ax = plt.subplot(6,2,plotnumber)
        sns.stripplot(y,x[column])
    plotnumber+=1
plt.tight_layout()
```

```python
#Great Lets proceed by checking multicollinearly in the dependent variables.Before that,we should 
# scale our data,Lets use the standard scalar for that.
```
