```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
```

```python
plt.figure(figsize=(10,15),facecolor='white')
plotnumber = 1

for column in data:
    if plotnumber <=17:
        ax = plt.subplot(9,2,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=30)
        
    plotnumber+=1
plt.tight_layout()
```

```python
# Lets rename the column names

data.rename(columns={"W":"wins","R":"runs","AB":"at bat","H":"hits","2B":"doubles","3B":"triples","HR":"homeruns","BB":"base on balls",
                    "SB":"stolen base","RA":"runs allowed","ER":"earned runs","ERA":"earned runs average","CG":"complete games",
                    "SHO":"shoutouts","SV":"save","E":"errors"},inplace=True)

print(data.columns)
```

```python
# Lets change the data in features and target.
y = data['wins']
x = data.drop(columns = ['wins'])
```

```python
plt.figure(figsize = (15,15),facecolor = 'white')
plotnumber = 1

for column in x:
    if plotnumber <=17:
        ax = plt.subplot(9,2,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('wins',fontsize=20)
    plotnumber +=1
plt.tight_layout()
```

```python
scaler = StandardScaler()

x_scaled = scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.25,random_state=21)
y_train.head()
```

```python
regression = LinearRegression()
regression.fit(x_train,y_train)
```

```python
# Adjusted R2 score
regression.score(x_train,y_train)
```

```python
#Lets check how model fits to test data.
regression.score(x_test,y_test)
```

```python
y_pred = regression.predict(x_test)
```

# Regularization

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings('ignore')
from sklearn.linear_model import Lasso
```

```python



parameters = { 'alpha':[.0001,.001,.01,.1,1,10],'random_state':list(range(0,10))}

ls = Lasso()
clf = GridSearchCV(ls,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
from sklearn.metrics import r2_score
ls = Lasso(alpha=1,random_state=0)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
pred_ls=ls.predict(x_test)

lss = r2_score(y_test,pred_ls)
lss
```

```python
cv_score=cross_val_score(ls,x,y,cv=5)
cv_mean=cv_score.mean()
cv_mean
```

# Ensemble Technique

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

parameters = {'criterion':['mse','mae'],'max_features':["auto","sqrt","log2"]}
rf = RandomForestRegressor()
clf = GridSearchCV(rf,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
rf=RandomForestRegressor(criterion='mse',max_features='auto')
rf.fit(x_train,y_train)
rf.score(x_train,y_train)
pred_decision = rf.predict(x_test)

rfs = r2_score(y_test,pred_decision)
print('R2 Score:',rfs*100)

rfscore= cross_val_score(rf,x,y,cv=5)
rfc=rfscore.mean()

print('Cross Val Score :',rfc*100)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
```

```python
parameters = {'criterion':['mse', 'friedman_mse', 'mae', 'poisson'],'max_features':["auto","sqrt","log2"]}
dt = DecisionTreeRegressor()
clf = GridSearchCV(dt,parameters)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
dt=DecisionTreeRegressor(criterion='friedman_mse',max_features='sqrt')
dt.fit(x_train,y_train)
dt.score(x_train,y_train)
pred_decision2 = dt.predict(x_test)

rfs2 = r2_score(y_test,pred_decision2)
print('R2 Score:',rfs*100)

rfscore2= cross_val_score(rf,x,y,cv=5)
rfc=rfscore2.mean()

print('Cross Val Score :',rfc*100)
```
